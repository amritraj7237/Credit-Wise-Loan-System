```python
# Create 1x2 subplot layout (since you have 2 graphs)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# ---- Gender Plot ----
gender_cnt = df["Gender"].value_counts()

sns.barplot(
    x=gender_cnt.index,
    y=gender_cnt.values,
    ax=axes[0]
)

axes[0].set_title("Gender Distribution")
axes[0].set_xlabel("Gender")
axes[0].set_ylabel("Count")

axes[0].bar_label(axes[0].containers[0])


# ---- Education Level Plot ----
education_level = df["Education_Level"].value_counts()

sns.barplot(
    x=education_level.index,
    y=education_level.values,
    ax=axes[1]
)

axes[1].set_title("Education Level Distribution")
axes[1].set_xlabel("Education Level")
axes[1].set_ylabel("Count")

axes[1].bar_label(axes[1].containers[0])


plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 2)

sns.boxplot(ax=axes[0, 0], data=df, x="Loan_Approved", y="Applicant_Income")
sns.boxplot(ax=axes[0, 1], data=df, x="Loan_Approved", y="Credit_Score")
sns.boxplot(ax=axes[1, 0], data=df, x="Loan_Approved", y="DTI_Ratio")
sns.boxplot(ax=axes[1, 1], data=df, x="Loan_Approved", y="Savings")

plt.tight_layout()
```
